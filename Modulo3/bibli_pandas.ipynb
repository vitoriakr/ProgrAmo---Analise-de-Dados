```python
dados.groupby('GENERO')['ID'].nunique()#agrupando o dataframe pela coluna 'GENERO' e contando o número de valores únicos na coluna 'ID' para cada grupo, o nunique é usado para contar o número de valores únicos, ou seja, o número de pessoas diferentes em cada grupo de gênero. O resultado é uma série onde o índice é o gênero e os valores são o número de pessoas únicas em cada gênero.                                                                        

#e o groupby é usado para agrupar os dados com base em uma ou mais colunas, neste caso, estamos agrupando por 'GENERO'. Depois de agrupar, podemos aplicar uma função de agregação, como 'nunique', para contar o número de valores únicos na coluna 'ID' para cada grupo de gênero.
```

```python
dados.groupby('GENERO', dropna=False)['ID'].nunique()#agrupando o dataframe pela coluna 'GENERO' e contando o número de valores únicos na coluna 'ID' para cada grupo, incluindo os valores nulos (NaN) como um grupo separado. O parâmetro dropna=False é usado para garantir que os valores nulos sejam incluídos no agrupamento. O resultado é uma série onde o índice é o gênero (incluindo NaN) e os valores são o número de pessoas únicas em cada gênero, incluindo os nulos.
```

```python
dados['GENERO'].value_counts(dropna=False)#contando o número de ocorrências de cada valor único na coluna 'GENERO', incluindo os valores nulos (NaN) como um valor separado. O resultado é uma série onde o índice é o gênero (incluindo NaN) e os valores são a contagem de ocorrências de cada gênero, incluindo os nulos. O parâmetro dropna=False é usado para garantir que os valores nulos sejam incluídos na contagem.
```

```python
dados[dados['IDADE'] > 30]['NIVEL'].value_counts()
```

```python
dados[(dados['IDADE'] > 30) & (dados['GENERO'] == 'Feminino')]['NIVEL'].value_counts()# vemos a contagem de cada nível de escolaridade para as pessoas do gênero feminino com idade maior que 30 anos. O filtro é aplicado primeiro para selecionar as linhas onde a coluna 'IDADE' é maior que 30 e a coluna 'GENERO' é igual a 'Feminino'. Em seguida, a função value_counts() é usada para contar o número de ocorrências de cada valor único na coluna 'NIVEL' para esse subconjunto de dados. O resultado é uma série onde o índice é o nível de escolaridade e os valores são a contagem de ocorrências de cada nível de escolaridade para as pessoas do gênero feminino com idade maior que 30 anos.
```

```python
pd.pivot_table(dados, values=['ID'], index=['GENERO'], columns=['GESTOR?'], aggfunc='count')#criando uma tabela dinâmica (pivot table) a partir do dataframe 'dados'. A tabela é organizada com 'GENERO' como índice (index), 'NIVEL' como colunas (columns) e os valores são a contagem de valores únicos na coluna 'ID' para cada combinação de gênero e nível de escolaridade. A função de agregação usada é 'nunique', que conta o número de valores únicos. O parâmetro fill_value=0 é usado para preencher os valores ausentes (NaN) com 0, indicando que não há pessoas únicas para aquela combinação específica de gênero e nível de escolaridade. O resultado é um DataFrame onde as linhas representam os gêneros, as colunas representam os níveis de escolaridade e os valores são a contagem de pessoas únicas em cada combinação.
```
