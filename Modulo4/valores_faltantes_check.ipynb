```python
import pandas as pd

dados = pd.read_csv("licitacoes.csv", encoding="latin1", sep=";")

```

```python
# 2. Identificar valores faltantes
print("Valores faltantes por coluna:")
print(dados.isnull().sum())
```

```python
# 3. Identificar a coluna com valores faltantes
coluna = "Data Abertura"  # você já descobriu qual é
# 4. Ver o tipo de dado da coluna
print("\nTipo de dado da coluna:")
print(dados[coluna].dtype)
```

```python
# 5. Converter para data
dados[coluna] = pd.to_datetime(dados[coluna], format="%d/%m/%Y", errors="coerce")

# 6. Substituir valores faltantes por uma data fictícia
dados[coluna] = dados[coluna].fillna(pd.Timestamp("1900-01-01"))

# 7. Conferir se ainda existem valores faltantes
print("\nValores faltantes após tratamento:")
print(dados.isnull().sum())
```
