```python
import matplotlib.pyplot as plt
import pandas as pd

# converter para número (segurança)
dados['SALARIO'] = pd.to_numeric(dados['SALARIO'], errors='coerce')

# gráfico está errado, não sei pq ?
plt.boxplot(dados['SALARIO'])
plt.title("Boxplot dos Salários")
plt.show()
```

```python
Q1 = dados['SALARIO'].quantile(0.25)
Q1#calculando o primeiro quartil da coluna SALARIO do dataframe
```

```python
Q3 = dados['SALARIO'].quantile(0.75)
Q3#calculando o terceiro quartil da coluna SALARIO do dataframe
```

```python
IQR = Q3 - Q1#calculando o intervalo interquartilico da coluna SALARIO do dataframe
IQR
```

```python
limite_superior = Q3 + (1.5 * IQR)  #calculando o limite superior para detectar outliers
limite_superior
```

```python
limite_inferior = Q1 - (1.5 * IQR)#calculando o limite inferior para detectar outliers
limite_inferior
```

```python
media_salario = dados['SALARIO'].mean()#calculando a media da coluna SALARIO do dataframe
media_salario
desvio_salario = dados['SALARIO'].std()#calculando o desvio padrao da coluna SALARIO do dataframe
desvio_salario
```

```python
limite_superior = media_salario + 3* desvio_salario#calculando o limite superior para detectar outliers
limite_superior
```

FEATURING ENGINEERGIN(adição de variaveis)

```python
def preencher_nivel(gestor, nivel):
    if gestor ==1:
        return 'Pessoa gestora'
    else:
        return nivel
```
