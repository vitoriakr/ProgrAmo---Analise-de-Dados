Estatistica Básica

```python
import numpy as np
#criando um array de numeros aleatorios
lista_idades = [26, 30, 32, 22, 26, 35, 40, 20, 43, 31, 23]
np.sum(lista_idades)#somando os valores da lista
```

```python
len(lista_idades)#contando o numero de elementos da lista
```

```python
np.sum(lista_idades)/len(lista_idades)#calculando a media das idades
```

```python
media =np.mean(lista_idades)#calculando a media da lista
print(f'Média aritmetica: {media}')
```

```python
lista_idades.sort()#ordenando a lista
lista_idades
```

```python
mediana = np.median(lista_idades)#calculando a mediana da lista
print(f'Mediana: {mediana}')
```
